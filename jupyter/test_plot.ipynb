```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({"GB":[1.31,1.67,2.63, 3.43, 3.3], "date":["0509", "0516", "0523", "0604", "0612"]})
# 将日期转换为正确的格式（添加20前缀表示年份）
df["date"] = "25" + df["date"]  # 变成 "200509", "200516"k等

# 转换为datetime类型
df["date"] = pd.to_datetime(df["date"], format="%y%m%d")
```

```python

# 绘制基本图形
ax = df.plot(x="date", y="GB", kind="line", 
             title="GB", figsize=(10, 5), 
             color="blue", legend=False,ylim=(1, 4))
# 设置x轴格式 - 只显示月-日
from matplotlib.dates import DateFormatter

date_format = DateFormatter("%m%d")
ax.xaxis.set_major_formatter(date_format)
# 添加数据值标签
for x, y in zip(df["date"], df["GB"]):
    ax.text(x, y, f'{y:.1f}', ha='center', va='bottom')

plt.show()
```
