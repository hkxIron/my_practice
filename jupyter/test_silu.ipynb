```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sigmoid(x) -> float:
    return 1/(1+np.exp(-x))
def silu(x):
    return x*sigmoid(x)
def dsigmoid(x):
    y=sigmoid(x)
    return y*(1-y)
def dsilu(x):
    sig = sigmoid(x)
    return sig +x*dsigmoid(x)
def relu(x):
    return np.where(x<0,0,x)

def drelu(x):
    return np.where(x<0,0,1)
```

```python
x=np.arange(-3, 3, 0.01)
y_sigmoid=sigmoid(x)
y_silu=silu(x)
y_dsigmoid=dsigmoid(x)
y_dsilu=dsilu(x)
y_relu = relu(x)
y_drelu = drelu(x)
```

```python
plt.plot(x, y_sigmoid, label='sigmoid')
plt.plot(x, y_silu, label='silu')
plt.plot(x, y_dsigmoid, label='dsigmoid')
plt.plot(x, y_dsilu, label='dsilu')
#plt.plot(x, y_relu, label='relu')
#plt.plot(x, y_drelu, label='drelu')
plt.legend()
plt.savefig('act_fun.svg')
```
